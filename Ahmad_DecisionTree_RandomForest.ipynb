```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
x=data.iloc[:,2:-1].values
y=data.iloc[:,-1].values
data
```

```python
x
```

```python
y
```

# train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
x_train
```

# feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

# decision tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=2)
classifier.fit(x_train,y_train)
```

# predicting new result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
y_pred=classifier.predict(x_test)
```

# confusion matrix & acuraccy

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Random forest

```python
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='entropy',random_state=42)
rfc.fit(x_train,y_train)
# Evaluation
rfc_pred_train=rfc.predict(x_train)
print('F1 score>',f1_score(y_train,rfc_pred_train))
```

# predicting new result

```python
rfc_pred_test=rfc.predict(x_test)
print('F1 score>',f1_score(y_test,rfc_pred_test))
```

```python
rfc.feature_importances_
```

```python
importances=pd.DataFrame(rfc.feature_importances_,columns=["imp"])
importances.sort_values(by='imp',ascending=False,inplace=True)
importances
```

```python
sns.displot(importances, x='imp', bins=2,shrink=0.5)
```
